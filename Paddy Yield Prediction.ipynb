```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('book1.csv')
```

```python
print(df)
```

```python
data = df[['Paddy variety','Taluks','Soil','Rainfall','Temperature','nitrogen','soil ph','Duration','Yield']]
print(data)
```

```python
import sklearn
from sklearn import linear_model
from sklearn import preprocessing
```

```python
model = preprocessing.LabelEncoder()
data['Soil'] = model.fit_transform((data['Soil']))
data['Taluks'] = model.fit_transform((data['Taluks']))
data['Paddy variety'] = model.fit_transform((data['Paddy variety']))
print(data)
```

```python
from mlxtend.plotting import scatterplotmatrix
cols = ["Rainfall","Temperature","nitrogen","soil ph","Duration","Yield"]
scatterplotmatrix(data[cols].values,figsize=(10,8),names = cols,alpha = 0.6)
plt.tight_layout()
plt.show()
```

```python
from mlxtend.plotting import heatmap
cm = np.corrcoef (data[cols].values.T)
hm = heatmap(cm,row_names = cols,column_names = cols)
plt.show()
```

```python
x = data.iloc[:, :-1]
y = data.iloc[:,8]
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state = 0)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
from sklearn.linear_model import Lasso
```

```python
lasso = Lasso(alpha = 1.0)
```

```python
lasso.fit(x_train,y_train)
```

```python
y_pred = lasso.predict(x_test)
```

```python
y_pred 
```

```python
print("Slope: %.2f"% lasso.coef_[0])
```

```python
print("Intercept: %.2f"% lasso.intercept_)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test,y_pred)
```

```python
np.sqrt(mean_squared_error(y_test,y_pred))
```

```python
n = ("Rsquared: %.2f"% lasso.score(x_test,y_test) )
print (n)
```

```python
model_diff = pd.DataFrame({'Yield' : y_test, 'Estimated yield': y_pred})
model_diff
```
